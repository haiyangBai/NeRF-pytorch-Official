```python
import torch
from torch import nn
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-6, 6, 100)

y_sigomid   = nn.Sigmoid()(x)
y_tanh      = nn.Tanh()(x)
y_relu      = nn.ReLU()(x)
y_softplus  = nn.Softplus()(x)

plt.figure(figsize=(18, 4))
plt.subplot(1, 4, 1)
plt.plot(x.data.numpy(), y_tanh.data.numpy(), 'r-')
plt.title('Tanh')
plt.grid()

plt.subplot(1, 4, 2)
plt.plot(x.data.numpy(), y_relu.data.numpy(), 'r-')
plt.title('ReLu')
plt.grid()

plt.subplot(1, 4, 3)
plt.plot(x.data.numpy(), y_softplus.data.numpy(), 'r-')
plt.title('Softplus')
plt.grid()

plt.subplot(1, 4, 4)
plt.plot(x.data.numpy(), y_sigomid.data.numpy(), 'r-')
plt.title('Sigmord')
plt.grid()

plt.show()
```
